Datasheet 4.01B: Migration to and from the United Kingdom by citizenship group, sex and age by country of last or next residence. Countries of last or next residence in groups."

```python
from databaker.framework import *
import pandas as pd
```

```python
from pathlib import Path

sourceFolder = Path('in')
sourceFolder.mkdir(exist_ok=True)

%run lib/scrape_ons.ipynb

metadata = scrape('https://www.ons.gov.uk/peoplepopulationandcommunity/populationandmigration/' \
                  'internationalmigration/datasets/ipscitizenshipgroupbysexbyagebycountryoflastornextresidence')
metadata
```

```python
inputFile = sourceFolder / 'data.xls'
response = session.get(metadata['fileURL'])
# with open(inputFile, 'wb') as f:
#   f.write(response.content)
# # tab = loadxlstabs(inputFile, sheetids='Datasheet 4.01A')[0]
```

```python
response.content
```

```python
xls = pd.ExcelFile(BytesIO(response.content))
df1 = pd.read_excel(xls, 'Datasheet 4.01B', header = 7)
```

```python
df1
```

```python
observations = df1.iloc[:, 5:]
observations.head()
```

```python
new_table = pd.melt(observations,
                       ["Row Label"],
                       var_name="Column label",
                       value_name="Value")
```

```python
new_table.head(50)
```

```python
new_table.shape
```

```python
new_table.dropna(inplace=True)
```

```python
new_table = new_table[new_table['Value'] != '.']
```

```python
new_table = new_table[new_table['Value'] != '0']
```

```python
new_table.head(50)
```

```python
new_table.shape
```

```python
new_table['country'] = new_table['Column label'].replace(to_replace= r"RESC (.*)(EST|CI )$", value = "\\1", regex=True)
new_table['measure'] = new_table['Column label'].replace(to_replace= r".* (EST|CI )$", value = "\\1", regex=True)
new_table['country'] = new_table['country'].str.lower()
new_table['country'] = new_table['country'].str.strip()
new_table['measure'] = new_table['measure'].str.lower()
new_table['measure'] = new_table['measure'].str.strip()
new_table['country'] = new_table['country'].replace(to_replace= r" ", value = "-", regex=True)
new_table['codelist'] = new_table['Column label'].str[0:4]
new_table['codelist'] = new_table['codelist'].str.lower()
new_table['Row Label'] = new_table['Row Label'].str.lower()
temp_table = new_table['Row Label'].str.split(',', -1, expand=True)
new_table['migration_flow'] = temp_table[0]
new_table['citizenship'] = temp_table[1]
new_table['sex'] = temp_table[2]
new_table['age'] = temp_table[3]
new_table['migration_flow'] = new_table['migration_flow'].str.strip()
new_table['migration_flow'] = new_table['migration_flow'].replace(to_replace= r" ", value = ":", regex=True)
new_table['citizenship'] = new_table['citizenship'].str.strip()
new_table['citizenship'] = new_table['citizenship'].replace(to_replace= r" ", value = ":", regex=True)
new_table['sex'] = new_table['sex'].str.strip()
new_table['age'] = new_table['age'].str.strip()
new_table['age'] = new_table['age'].replace(to_replace= r" ", value = ":", regex=True)
new_table['Geography'] = 'K02000001'
new_table['Unit'] = 'People (thousands)'
new_table['Measure Type'] = 'Count'
new_table['Year'] = '2016'
temp_table = new_table[:]
temp_table['Keys'] = temp_table['Row Label'] + '/' + temp_table['Column label']
temp_table['Keys'] = temp_table['Keys'].str.strip()
temp_table['Keys2'] = temp_table['Keys'].str.rstrip('EST')
temp_table['Keys2'] = temp_table['Keys2'].str.rstrip('CI')
temp_table['Keys2'] = temp_table['Keys2'].str.strip()
Final_table = pd.merge(temp_table, temp_table[['Keys2','Value','measure']], how = 'left', left_on = 'Keys2', right_on = 'Keys2')
Final_table = Final_table[Final_table['measure_x'] != Final_table['measure_y']]
Final_table = Final_table.drop_duplicates('Keys2', keep = 'first')
Final_table.columns = ['Value' if x=='Value_x' else x for x in Final_table.columns]
Final_table.columns = ['CI' if x=='Value_y' else x for x in Final_table.columns]
Final_table['Value'] = Final_table['Value'].astype(float)
Final_table['Year'] = Final_table['Year'].astype(int)
Final_table['CI'] = Final_table['CI'].astype(float)
Final_table = Final_table[['Geography','Year','codelist','country','migration_flow','citizenship','sex','age','Measure Type','Value','CI','Unit']]
```

```python
destinationFolder = Path('out')
destinationFolder.mkdir(exist_ok=True, parents=True)

Final_table.to_csv(destinationFolder / ('tidydata4.01B.csv'), index = False)
```
