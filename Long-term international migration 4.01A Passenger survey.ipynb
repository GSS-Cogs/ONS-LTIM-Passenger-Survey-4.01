```python
from databaker.framework import *
import pandas as pd
```

```python
from pathlib import Path

sourceFolder = Path('in')
sourceFolder.mkdir(exist_ok=True)

%run lib/scrape_ons.ipynb

metadata = scrape('https://www.ons.gov.uk/peoplepopulationandcommunity/populationandmigration/' \
                  'internationalmigration/datasets/ipscitizenshipgroupbysexbyagebycountryoflastornextresidence')
metadata
```

```python
inputFile = sourceFolder / 'data.xls'
response = session.get(metadata['fileURL'])
# with open(inputFile, 'wb') as f:
#   f.write(response.content)
# # tab = loadxlstabs(inputFile, sheetids='Datasheet 4.01A')[0]
```

```python
response.content
```

```python
xls = pd.ExcelFile(BytesIO(response.content))
df1 = pd.read_excel(xls, 'Datasheet 4.01A', header = 7)
```

```python
df1
```

```python
observations = df1.iloc[:, 5:]
observations.head()
```

```python
new_table = pd.melt(observations,
                       ["Row Label"],
                       var_name="Column label",
                       value_name="Value")
```

```python
new_table.head(50)
```

```python
new_table.shape
```

```python
new_table.dropna(inplace=True)
```

```python
new_table = new_table[new_table['Value'] != '.']
```

```python
new_table = new_table[new_table['Value'] != '0']
```

```python
new_table.head(50)
```

```python
sourceFilename = 'column-labels-revised.csv'
sourceFile = sourceFolder / sourceFilename
sourceUrl = 'https://drive.google.com/uc?export=download&id=1G3tiDCLxNxyrQ95W94DXsHNMqBxv45NX'

if not (sourceFile.exists() and sourceFile.is_file()):
    response = requests.get(sourceUrl)
    with open(sourceFile, 'wb') as f:
        f.write(response.content)

columns = pd.read_csv(sourceFile)
```

```python
sourceFilename = 'row-labels-revised.csv'
sourceFile = sourceFolder / sourceFilename
sourceUrl = 'https://drive.google.com/uc?export=download&id=1baA0s5L1PdbK0VPWgNde70yjGFSAeZpa'

if not (sourceFile.exists() and sourceFile.is_file()):
    response = requests.get(sourceUrl)
    with open(sourceFile, 'wb') as f:
        f.write(response.content)

rows = pd.read_csv(sourceFile)
```
